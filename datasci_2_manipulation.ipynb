#Identifying Missing Data
The missing data in this datafile isn't actually considered to be missing, but is considered a string called "missing". This code allows you to see every column that contains the string, "missing". Using the pandas.DataFrame.apply, with the index of axis = 1, we can append more functions to the dataframe. Axis = 1 refers to the column, allowing us to apply the function to each row. By using a lambda function, we can apply the function to each element of the series. Using pandas.DataFrame.astype, we can apply functions to a specific data type, and in our case, we used strings. Case being set to false allows us to ignore being case sensitive.

```python
substring = "missing"
df[df.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)]
```

This code allows us to sum up how many times the string "missing" appears in each column. Pandas.dataframe.eq allows us to compare each element in the dataframe to the string "missing. We then append the sum function. The second part of the code uses a list of the m variable we specified before, and the values of each column as a numpy array. Columns=m allows us to set the column name for the new dataframe. m.index provides the column name from the m variable.

```python
m=df.eq('missing').sum()
pd.DataFrame([m.values],columns=m.index)
```

#Removing White Space and Special Characters

This code created a function called cleaning. Within this function is another function called clean_name, which we will pass one argument being 'name' which will be the column name. Then, cleaned_name uses the re.sub function, which replaces the strings annoted inside. They are replaced with an empty string. The .lower converts the cleaned name to lower case. The next line of code, df.columns allows us to pass through the columns attribute with the functions we created before.

```python
def cleaning(df):
    def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

    df.columns = [clean_name(col) for col in df.columns]
    return df

df = cleaning(df)
df

```

#Transforming Data

##Checking if tachycardia

This code created a function called check_tachycardia, passing through the variable x. x is set to convert to a float. "Yes" is returned if x >= high_hr, which is set to 100, if notm it'll return "No". Results in ValueError are also returned for "no" to account for non-numberic values or strings. Then, we apply the function on our dataframe, df.

```python
high_hr = 100

def check_tachycardia(x):
    try:
        x = float(x)  # Try to convert to a float
        return "Yes" if x >= high_hr else "No"
    except ValueError:
        return "No"  # Return "No" for non-numeric values

df["tachycardia"] = df["averageheartrate"].apply(check_tachycardia)
df
```

##Converting height in centimeters to inches

This code converts the height in cm to height in inches. It is similar to the checking if tachycardia code.

```python
def cmtoin(x):
  try:
    x = float(x)
    return x/2.54 if x else "Unavailable"
  except ValueError:
    return "Unavailable"

df["heightininches"] = df["heightincm"].apply(cmtoin)
df
```

##Aggregating Data

This code groups average blood pressure and aggregates the mean of heights in inches. This will see the average height listed for various blood pressure rates.

```python
df.groupby("averagebp").agg('mean')
```

This code removes the NaN values and replaces them with None.

```python
df.dropna()
```

This code looks at the averagebp and lists the average of height in inches as well as the sum of the duration of visit duration in minutes.

```python
df.groupby("averagebp").agg(
        meanheight=pd.NamedAgg(column="heightininches", aggfunc="mean"),
    sumduration=pd.NamedAgg(column="visitdurationmins", aggfunc="sum")
)
```

This code creates a pivot table using the index of the state of residence. The column is the number of tests. The value takes the sum of the visit durations.

```python
pivot_data = df.pivot_table(index='stateofresidence', columns='numberoftests', values='visitdurationmins', aggfunc='sum')
pivot_data
```
