#Identifying Missing Data
The missing data in this datafile isn't actually considered to be missing, but is considered a string called "missing". This code allows you to see every column that contains the string, "missing". Using the pandas.DataFrame.apply, with the index of axis = 1, we can append more functions to the dataframe. Axis = 1 refers to the column, allowing us to apply the function to each row. By using a lambda function, we can apply the function to each element of the series. Using pandas.DataFrame.astype, we can apply functions to a specific data type, and in our case, we used strings. Case being set to false allows us to ignore being case sensitive.

```python
substring = "missing"
df[df.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)]
```

This code allows us to sum up how many times the string "missing" appears in each column. Pandas.dataframe.eq allows us to compare each element in the dataframe to the string "missing. We then append the sum function. The second part of the code uses a list of the m variable we specified before, and the values of each column as a numpy array. Columns=m allows us to set the column name for the new dataframe. m.index provides the column name from the m variable.

```python
m=df.eq('missing').sum()
pd.DataFrame([m.values],columns=m.index)
```

#Replacing "missing" with NaN

This code replaces a string with NaN, or actual missing values. We then used the pandas.DataFrame.replace function. We then use the numpy function to replace a value with NaN. inplace=True lets us modify our dataframe directly.

```python
df.replace('missing', pd.np.nan, inplace=True)
df
```

The next line of code, df.columns allows us to pass through the columns attribute with the functions we created before.
